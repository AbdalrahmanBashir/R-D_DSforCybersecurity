```python
#Subsets 
#Crude way
crude_subset_malign = []
for each_attack in data_table:
    if each_attack['label'] == 1 and each_attack['port_no']==3:
        ##print(each_attack)
        crude_subset_malign.append(each_attack)
print(len(crude_subset_malign))
```

```python
#Activity two extend The above code
new_subset = []
for each_attack in data_table:
    if each_attack['label'] == 0 and each_attack['port_no']==2 and each_attack['flows']==3:
        ##print(each_attack)
        new_subset.append(each_attack)
print(len(new_subset))
```

```python
print(new_subset)
```

```python
#Semi-fancy way
from collections import Counter
arra_malign = []
array_malign = [each_attack['label'] == 1 and each_attack['port_no'] == 3 for each_attack in data_table]
dict(Counter(array_malign))
```

```python
#Activity three extend the above Semi-fancy way
arra_malign_two = []
array_malign_two = [each_attack['label'] == 0 and each_attack['port_no'] == 2 and each_attack['flows']==3 for each_attack in data_table]
dict(Counter(array_malign_two))
```

```python
#Fancy way
good_subset = []
good_subset = data_table[np.where((data_table['label']!=1) & (data_table['port_no']!=3))]
```

```python
not_attack_subset = []
not_attack_subset = data_table[np.where((data_table['label']!=1)& (data_table['port_no']!=3))]
fig =  plt.figure()
plt.hist(not_attack_subset['rx_bytes']-not_attack_subset['tx_kbps'], bins=5)
plt.show() # this will tell the python environment to show the plot.
```

```python
#Activity one # please report the output of histogram plot it look informative?

plt.hist(not_attack_subset['rx_bytes']-not_attack_subset['tx_kbps'], bins=50, color='red',histtype='step', label='no attack')
plt.xlabel('(rx_bytes-tx_kbps) color')
plt.show()
```
