```python
import numpy as np
import matplotlib.pyplot as plt
# Load the data
data_table = np.genfromtxt("./Denial-of-Service.csv", delimiter=",", names=True)
# Select two columns
col1 = "dur"
col2 = "dur_nsec"
# Sort the data based on col1
sorted_indices = np.argsort(data_table[col1])
sorted_col1 = data_table[col1][sorted_indices]
sorted_col2 = data_table[col2][sorted_indices]
# Create the scatter plot
plt.figure(figsize=(7, 5))
plt.scatter(data_table[col1], data_table[col2], marker="s", color='blue', label=f"{col1} vs. {col2}")
# Add labels and title
plt.xlabel(col1)
plt.ylabel(col2)
plt.title(f"Scatter Plot of {col1} vs. {col2}")
# Connect scatterplot points with a solid line
plt.plot(sorted_col1, sorted_col2, linestyle='solid', color='green', label='Connected Points')

# Add legend
plt.legend()

# Show the plot
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Load the data
data_table = np.genfromtxt("Denial-of-Service.csv", delimiter=",", names=True)

# Select columns
col1 = "pktperflow"
col2 = "byteperflow"

# Sort the data based on col1
sorted_indices = np.argsort(data_table[col1])
sorted_col1 = data_table[col1][sorted_indices]
sorted_col2 = data_table[col2][sorted_indices]

# Create the scatter plot with square marker
plt.scatter(data_table[col1], data_table[col2], marker="s", color='blue', label=f"{col1} vs. {col2}")


# Add labels and title
plt.xlabel(col1)
plt.ylabel(col2)
plt.title(f"Scatter Plot of {col1} vs. {col2} with Solid Line and Square Marker")

# Connect scatterplot points with a solid line
plt.plot(sorted_col1, sorted_col2, linestyle='-', color='green', alpha=0.5, label='Connected Points')

# Add legend
plt.legend()

# Show the plot
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate random 2D data
x = np.random.randn(50)
y = np.random.randn(50)

# Create a 2D histogram
plt.hist2d(x, y, bins=(10, 10), cmap='RdBu')

# Add labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('2D Histogram Example')

# Add a colorbar
cb = plt.colorbar()
cb.set_label('Frequency')

# Add a scatter plot on top of the 2D histogram
plt.scatter(x, y,s=50, marker='.', alpha=0.5, label='Data Points')

# Add a legend
plt.legend()

# Save the figure to a location
plt.savefig('2Dhistogram.png')

# Show the plot
plt.show()
```
